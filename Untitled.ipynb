# Рекомендательная система

```python
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
```

```python
df = pd.read_excel("Online Retail.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.loc[df["Quantity"]>0]
```

```python
df = df.dropna(subset=["CustomerID"])
```

```python
customer_id_matrix = df.pivot_table(index = "CustomerID", columns = "StockCode", values = "Quantity", aggfunc='sum' )
customer_id_matrix.head()
```

```python
customer_id_matrix.shape
```

```python
df["CustomerID"].nunique(), df['StockCode'].nunique()
```

```python
customer_id_matrix = customer_id_matrix.applymap(lambda x: 1 if x>0 else 0)
```

```python
customer_id_matrix.head()
```

```python
m = customer_id_matrix.values
m
```

```python
def cos_sim(array):
    csr_array = csr_matrix(array)
    numerator = np.dot(csr_array,csr_array.T).toarray()
    denomenator = np.linalg.norm(array,axis=1,keepdims=True)*np.linalg.norm(array.T,axis=0,keepdims=True)
    return numerator/denomenator
```

```python
user_user_sim_matrix = cos_sim(m)
user_user_sim_matrix
```

```python
user_user_sim_matrix = pd.DataFrame(user_user_sim_matrix)
user_user_sim_matrix.columns = customer_id_matrix.index
user_user_sim_matrix['CustomerID'] = customer_id_matrix.index
user_user_sim_matrix
```

```python
user_user_sim_matrix = user_user_sim_matrix.set_index('CustomerID')
```
